```python
# import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the CVD_cleaned dataset
df = pd.read_csv("CVD_cleaned.csv")
```

```python
#see the elements of the dataset
df.head()
```

```python
# Get a general overview of the dataset
df.info()
```

```python
# Create summary statistics
df.describe()
```

```python
# Check for missing values
df.isna().sum()
```

```python
# Print the column names
print(df.columns)
```

```python
# Explore the distributions of the variables
df.hist(figsize=(10, 10))
plt.subplots_adjust(bottom=0.15)
plt.show()
```

```python
# Look for relationships between variables
sns.pairplot(df)
plt.show()
```

```python
# Create a scatter plot of self-reported general health status vs. heart disease
sns.scatterplot(
    x="General_Health",
    y="Heart_Disease",
    data=df,
)
plt.show()
```

```python
# Create a bar plot that shows the count of heart disease cases for each age category.
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Age_Category', hue='Heart_Disease')
plt.title('Bar Plot of Heart Disease by Age Category')
plt.xlabel('Age Category')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()
```

```python
# Create a bar plot that displays the distribution of alcohol consumption levels (e.g., 'Low', 'Moderate', 'High') among individuals with and without heart disease.
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Alcohol_Consumption', hue='Heart_Disease')
plt.title('Bar Plot of Heart Disease by Alcohol Consumption')
plt.xlabel('Alcohol Consumption')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()
```

```python
# Group the data by 'Age_Category' and calculate the average weight for each group
avg_weight_by_age = df.groupby('Age_Category')['Weight_(kg)'].mean().reset_index()

# Create a line plot of the average weight over different age categories
plt.figure(figsize=(8, 6))
sns.lineplot(data=avg_weight_by_age, x='Age_Category', y='Weight_(kg)', marker='o')
plt.title('Average Weight by Age Category')
plt.xlabel('Age Category')
plt.ylabel('Average Weight (kg)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.show()
```

```python
#Histogram for a age distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['Age_Category'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age_category')
plt.ylabel('Count')
plt.show()
```

```python
# Create a histogram of BMI
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='BMI', bins=20, kde=True)
plt.title('BMI Distribution')
plt.xlabel('BMI')
plt.ylabel('Count')
plt.show()
```

```python
# Create a count plot that shows the distribution of individuals with and without heart disease
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Heart_Disease')
plt.title('Distribution of Heart Disease')
plt.xlabel('Heart Disease (0: No, 1: Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# Create a count plot that shows the distribution of individuals with and without diabetes 
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Diabetes')
plt.title('Distribution of Diabetes')
plt.xlabel('Diabetes (0: No, 1: Yes)')
plt.ylabel('Count')
plt.show()
```

```python
#Box plot for BMI by Heart Disease presence
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Heart_Disease', y='BMI')
plt.title('BMI by Heart Disease Presence')
plt.xlabel('Heart Disease (0: No, 1: Yes)')
plt.ylabel('BMI')
plt.show()

```

```python
# Create a box plot that shows the distribution of weights ('Weight_(kg)') across different 'Age_Category' groups
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Age_Category', y='Weight_(kg)')
plt.title('Box Plot of Weight by Age Category')
plt.xlabel('Age Category')
plt.ylabel('Weight (kg)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Create a relplot to visualize the relationships between weight, height, age category, sex, and BMI
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.relplot(data=df, x='Weight_(kg)', y='Height_(cm)', hue='Age_Category', style='Sex', size='BMI', sizes=(10, 200))
plt.title('Relationship Between Weight, Height, and Age Category')
plt.xlabel('Weight (kg)')
plt.ylabel('Height (cm)')
plt.show()
```

```python
# Create a relplot to show relationship Between Alcohol, Fruit, and Green Vegetables Consumption
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.relplot(data=df, x='Alcohol_Consumption', y='Fruit_Consumption', hue='Green_Vegetables_Consumption', size='BMI', sizes=(10, 200))
plt.title('Relationship Between Alcohol, Fruit, and Green Vegetables Consumption')
plt.xlabel('Alcohol Consumption')
plt.ylabel('Fruit Consumption')
plt.show()
```

```python
# Count the occurrences of each category in the 'Heart_Disease' column
heart_disease_counts = df['Heart_Disease'].value_counts()

# Create a pie chart showing the distribution of heart disease
plt.figure(figsize=(6, 6))
plt.pie(heart_disease_counts, labels=heart_disease_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Heart Disease')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python

# Count the occurrences of each category in the 'Diabetes' column
diabetes_counts = df['Diabetes'].value_counts()

# Create a pie chart that displays the distribution of individuals with and without diabetes 
plt.figure(figsize=(6, 6))
plt.pie(diabetes_counts, labels=diabetes_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Diabetes')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Create a categorical plot (catplot) to examine the relationship between general health status and heart disease
plt.figure(figsize=(10, 6))
sns.catplot(data=df, x='General_Health', hue='Heart_Disease', kind='count')
plt.title('Relationship Between General Health Status and Heart Disease')
plt.xlabel('General Health Status')
plt.ylabel('Count')
plt.show()
```

```python
# Create a count plot to examine the relationship between medical checkups and heart disease
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Checkup', hue='Heart_Disease')
plt.title('Relationship Between Medical Checkups and Heart Disease')
plt.xlabel('Frequency of Medical Checkups')
plt.ylabel('Count')
plt.show()
```

```python
# Create a count plot to examine the relationship between smoking history and heart disease
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Smoking_History', hue='Heart_Disease')
plt.title('Relationship Between Smoking History and Heart Disease')
plt.xlabel('Smoking History')
plt.ylabel('Count')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()
```

```python
# Create a count plot to examine the relationship between alcohol consumption and heart disease
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Alcohol_Consumption', hue='Heart_Disease')
plt.title('Relationship Between Alcohol Consumption and Heart Disease')
plt.xlabel('Alcohol Consumption')
plt.ylabel('Count')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()
```
